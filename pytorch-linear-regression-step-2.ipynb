```python
import torch.nn as nn
# Define data
# Define the model in this case is the w*x linear regression
# Model prediction is the applications of the model wit h given parameters
# Learning loop
# Loss how far we are vs prediction
# Gradient calculation vs parameteres
# Weight update
# final results
```

```python
# Replacing np array for torch.tensor
#X =np.array([1,2,3,4],dtype=np.float32)
#Y= np.array([2,4,6,8],dtype=np.float32)

X = torch.tensor([1,2,3,4],dtype=torch.float32)
Y = torch.tensor([2,4,6,8],dtype=torch.float32)    
w = torch.tensor(0.0,dtype=torch.float32,requires_grad=True)
```

```python
print(X)
print(Y)
```

```python
#model prediction
def forward(x):
    return w*x
```

```python
#loss
def loss(y,y_predicted):
    return((y_predicted-y)**2).mean()
```

```python
#gradient = backwards pass
#replacing manual gradient for automatic one
#MSE = 1/N *(w*x-y)**2
#dj/dw = 1/N 2x (w*x-y)
#def gradient (x,y,y_predicted):
#    return np.dot(2*x,y_predicted-y).mean()

```

```python
print(f'prediction before training:f(5)={forward(5):.3f}')
```

```python
#training
learning_rate = 0.01
n_iters= 100
```

```python
for epoch in range (n_iters):
    #prediction  = forward pass
    y_pred = forward(X)
    #loss
    l = loss (Y,y_pred)
    #gradients
    l.backward() #dl/dw
    #update weights
    with torch.no_grad():
        w -= learning_rate * w.grad
    w.grad.zero_()
    if epoch % 10 ==0:
        print(f'{epoch+1}: w={w:.3f},loss={l:.8f}')
        
```

```python
print(f'prediction after training:f(5)={forward(5):.3f}')
```
